### Blue solid lines are the default for matplotlib so lets look into changing the graphs.

#### Other options for the color characters are:
#### 'r' = red
#### 'g' = green
####  'b' = blue
#### 'c' = cyan
#### 'm' = magenta
####  'y' = yellow
####  'k' = black
#### 'w' = white
#### Options for line styles are
####  '-' = solid
####  '--' = dashed
#### ':' = dotted
#### '-.' = dot-dashed
#### '.' = points
#### 'o' = filled circles
####  '^' = filled triangles

### Let's try another graph with magenta color and triangles.

```python
plt.plot([1,2,3,4,5,3.5], 'm^')
```

### Another feature is that you can set your own axis lets set x to be 10 and y to be 20.

```python
plt.plot([1,2,3,4,15,20,17,18,3] ,'r-')
plt.axis([0, 10, 0, 20])
```

### With the plt.axis you can set the variables for the y axis and x axis lets try setting 1,5 for x and 2,10 for y.

```python
plt.plot([4,2,7,6], 'g--')
plt.axis([1, 5, 2, 10])
```

### If you choose to plot two brackets of information that is the plot on the graph with the first values being x and the second values being y.

```python
plt.plot([2,4,7,8,2,3.5,7,8], [1,5,9,16,2,2.2,9,19], 'k.')
plt.axis([0, 20, 0, 20])
```

### Now that we understand the basics of how to graph lets move on to adding titles to the x and y axis by simply using plt.xlabel and plt.ylabel you can chose what to call the axis.

```python
plt.plot([1,2,3,4], [1,4,9,16], 'yo')
plt.axis([0, 6, 0, 20])
plt.xlabel('this is x!')
plt.ylabel('this is y!')
```

### To add more than one graph to be shown just execute the plot code again to place another line.

```python
plt.plot([1,2,3], 'yo')
plt.plot([4,5,6],'k.') 
plt.plot ([6,4,1],'g--')
plt.plot([2,5,2],'m-.')
plt.xlabel('Dollars')
plt.ylabel('Hours')
```

### With the plt.title you can name the graph and give it a title.

```python
plt.plot([1,2,3],'-b')
plt.plot([4,5,6],'-r') 
plt.title('My First Plot')
```

### With the label = you can label the graphs and adding the plt.legend to display them on the graph.

```python
plt.plot([1,2,3],'-b', label='Graph 1')
plt.plot([2,7,4],'-m', label = 'Graph 2') 
plt.plot([3,8,5],'-g', label = 'Graph 3') 
plt.plot([4,1,6],'-c', label = 'Graph 4') 
plt.plot([5,9,7],'-k', label = 'Graph 5') 
plt.title('Plot 2')
plt.xlabel('Dollars')
plt.ylabel('Hours')
plt.legend()
```

### There are many different locations to place the location of the legend just use  loc = 

#### right,  center left, upper right, lower right, best, center, lower left, center right, upper left,  upper center, lower center

  

```python
plt.plot([1,2,3],'-b', label='Graph 1')
plt.plot([2,7,4],'-m', label = 'Graph 2') 
plt.plot([3,8,5],'-g', label = 'Graph 3') 
plt.plot([4,1,6],'-c', label = 'Graph 4') 
plt.plot([5,9,7],'-k', label = 'Graph 5') 
plt.title('Plot 2')
plt.xlabel('Dollars')
plt.ylabel('Hours')
plt.legend(loc='best')
```

### Using the Right legend location.

```python
plt.plot([1,2,3],'-b', label='Graph 1')
plt.plot([2,7,4],'-m', label = 'Graph 2') 
plt.plot([3,8,5],'-g', label = 'Graph 3') 
plt.plot([4,1,6],'-c', label = 'Graph 4') 
plt.plot([5,9,7],'-k', label = 'Graph 5') 
plt.title('Plot 3')
plt.xlabel('Dollars')
plt.ylabel('Hours')
plt.legend(loc='right')
```

## lmplot

### With Seaborn lmplot will show a scatterplot with a trend line.

```python
sns.lmplot(x='AGE', y='AMOUNT', data=df)
```

### With the x_bins used it will throw the points into an amount of bins for this case 6.

```python
sns.lmplot(x='AGE', y='AMOUNT', data=df, x_bins=6)
```

### With the x_ci and ci used it will change the confidence interval for the graph.

```python
sns.lmplot(x='AGE', y='AMOUNT', data=df, x_bins=6, x_estimator=np.mean, x_ci=50, ci = 20)
```

### You can hue your graph by another variable. This will show what categories that these values fall into also with the fit_reg set to true it will add the regression line.

```python
sns.lmplot(x='AGE', y='AMOUNT', data=df, fit_reg=True, hue='JOB')  
```

### With the fit_reg set to false it will not create regression line.

```python
sns.lmplot(x='AGE', y='AMOUNT', data=df, fit_reg=False,  hue='JOB')   
```

### You can also set limits for the graph to only display certain values.

```python
sns.lmplot(x='AGE', y='AMOUNT', data=df, fit_reg=False, hue='JOB')
plt.ylim(20, 10000)
plt.xlim(20, 40)
```

### Making FacetGrids with random data.

```python
# Generate an example radial datast
r = np.linspace(0, 10, num=100)
l = pd.DataFrame({'r': r, 'slow': r, 'medium': 2 * r, 'fast': 4 * r})

# Convert the dataframe to long-form or "tidy" format
l = pd.melt(l, id_vars=['r'], var_name='speed', value_name='theta')

# Set up a grid of axes with a polar projection
g = sns.FacetGrid(l, col="speed", hue="speed",
                  subplot_kws=dict(projection='polar'), size=4.5,
                  sharex=False, sharey=False, despine=False)

# Draw a scatterplot onto each axes in the grid
g.map(plt.scatter, "theta", "r")

#1_datavisualization_overview_tutorial_s
```
